```python
import os
import statistics
import math
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
%matplotlib inline
```

```python
#this little function will help me making my code a little tidier.
def avg(x):
    a=statistics.mean(x)
    return a
```

```python
os.chdir('C:\\Users\\rafal\\Desktop\\Licencjat')
data=pd.read_csv('players_21.csv',sep=',',engine='python')
data
```

```python
age=pd.Series(['24-26','27-29', '30-32', '33-35'])

#getting proper age intervals
data1822=data[data['age']>=24]
data1822=data1822[data1822['age']<=26]
data2226=data[data['age']>=27]
data2226=data2226[data2226['age']<=29]
data2630=data[data['age']>=30]
data2630=data2630[data2630['age']<=32]
data3034=data[data['age']>=33]
data3032=data3034[data3034['age']<=35]

overall=pd.Series(np.transpose(np.r_[avg(data1822['overall']),avg(data2226['overall']),avg(data2630['overall']), avg(data3034['overall'])]))
movement_sprint_speed=pd.Series(np.transpose(np.r_[avg(data1822['movement_sprint_speed']),avg(data2226['movement_sprint_speed']),avg(data2630['movement_sprint_speed']), avg(data3034['movement_sprint_speed'])]))
movement_agility=pd.Series(np.transpose(np.r_[avg(data1822['movement_agility']),avg(data2226['movement_agility']),avg(data2630['movement_agility']), avg(data3034['movement_agility'])]))
power_stamina=pd.Series(np.transpose(np.r_[avg(data1822['power_stamina']),avg(data2226['power_stamina']),avg(data2630['power_stamina']), avg(data3034['power_stamina'])]))
skill_dribbling=pd.Series(np.transpose(np.r_[avg(data1822['skill_dribbling']),avg(data2226['skill_dribbling']),avg(data2630['skill_dribbling']), avg(data3034['skill_dribbling'])]))

outcome=pd.concat([overall,movement_sprint_speed,movement_agility,power_stamina,skill_dribbling],axis=1).astype(int)
outcome=pd.concat([age,outcome],axis=1)
outcome.columns=['Age','Overall rating', 'Sprint speed','Agility','Stamina','Dribbling']
outcome
```

```python
data1=data[data['age']<38]
data1=data1.groupby('age').mean()
data1
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 2))

ax[0].axis('off')
ax[0].axis('tight')
ax[0].table(cellText=outcome.values, colLabels=outcome.columns, loc='center')

ax[1].plot(data1['movement_sprint_speed'], label='Sprint speed')
ax[1].plot(data1['movement_agility'],label='Agility')
ax[1].plot(data1['power_stamina'],label='Stamina')
ax[1].plot(data1['skill_dribbling'],label='Dribbling')
ax[1].set_xlabel('age')
ax[1].set_ylabel('Average values')
ax[1].legend(loc='best')

plt.show()
```
